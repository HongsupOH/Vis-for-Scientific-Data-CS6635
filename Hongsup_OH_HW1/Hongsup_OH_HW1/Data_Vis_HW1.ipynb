```python
def genInc(n):
    """
    Function for creating 
    monotonic increaging bins
    
    bins = genInc(9)
    """
    arr = [1]
    dif = [0.4]
    for i in range(1,n-1):
        dif.append(dif[i-1]+0.1)
    ind = 0
    cur = arr[ind]
    for i in range(len(dif)):
        arr.append(cur+dif[i])
        ind+=1
        cur = arr[ind]
    
    return arr

bins = genInc(10)
plt.hist(x,bins,rwidth=0.2,facecolor='blue')
plt.xlabel('Bins')
plt.ylabel('Frequency')
```

```python
def genDec(n):
    """
    Function for creating 
    monotonic increaging bins
    
    bins = genDec(9)
    """
    arr = [1]
    dif = [0.4]
    for i in range(1,n-1):
        dif.append(dif[i-1]+0.1)
    dif = dif[::-1]
    ind = 0
    cur = arr[ind]
    for i in range(len(dif)):
        arr.append(cur+dif[i])
        ind+=1
        cur = arr[ind]
    return arr

bins1 = genDec(10)
plt.hist(x,bins1,rwidth=0.2,facecolor='blue')
plt.xlabel('Bins')
plt.ylabel('Frequency')
```

```python
f = open("NOAA-Temperatures.csv")

def genArrs(f):
    """
    Separate year data and degree data
    to two different arrays
    
    yr,value = genArrs(f)
    """
    lines = f.readlines()
    yr,value=[],[]
    for i in range(5,len(lines)):
        cur = lines[i].split(",")
        cur[1] = cur[1][:len(cur[1])-1]
        yr.append(float(cur[0]))
        value.append(float(cur[1]))
    yr = np.array(yr)
    value = np.array(value)
    return yr,value

yr,value = genArrs(f)
```

```python
Y = yr
V = value
plt.scatter(Y,V)
plt.xlabel('Year')
plt.ylabel('Degrees Celsius +/- From Average')
plt.grid()
```

```python
plt.bar(Y,V)
plt.xlabel('Year')
plt.ylabel('Degrees Celsius +/- From Average')
plt.grid()
```

```python
def movingAvg(arr,n):
    """
    In order to reduce fluctuation
    moving averge concept is used
    
    movAves = movingAvg(V,N)
    
    """
    cSum, movAves = [0],[]
    for i,num in enumerate(arr,1):
        cSum.append(cSum[i-1] + num)
        if i>=n:
            movAve = (cSum[i] - cSum[i-n])/n
            movAves.append(movAve)
    return movAves

N = 10
movAves = movingAvg(V,N)
plt.plot(Y,V)
plt.plot(Y[N-1:],movAves)
plt.legend(['Normal plot','Moving average'])
plt.xlabel('Year')
plt.ylabel('Degrees Celsius +/- From Average')
plt.grid()
```

```python
from collections import defaultdict
f1 = open('both_sexes.csv')

def genData(f1):
    """
    Generate hash map 
    for using star plot
    
    labels,data2 = genData(f1)
    """
    lines1 = f1.readlines()
    nl = lines1[0].split(",")
    labelsCandi = [x.replace('"', '') for x in nl][3:]
    data2 = defaultdict(list)
    for i in range(1,len(lines1)):
        cur = lines1[i].split(",")[1:]
        dt = []
        for x in cur[3:]:
            if x=="NA":
                dt.append(x)
            else:
                dt.append(float(x))
        data2[cur[1]]=dt
    return labelsCandi,data2

labelsCandi,data2 = genData(f1)
years = np.array(list(data2))
print(years)
```

```python
num = 10
labels = labelsCandi[:num]
stats = data2['2004-01-01'][:num]
stats1 = data2['2008-01-01'][:num]
stats2 = data2['2010-01-01'][:num]

originStat = stats[:]
originStat1 = stats1[:]
originStat2 = stats2[:]

angles = np.linspace(0, 2*np.pi,len(labels),endpoint=False)

stats = np.concatenate((stats,[stats[0]]))
stats1 = np.concatenate((stats1,[stats1[0]]))
stats2 = np.concatenate((stats2,[stats2[0]]))

angles = np.concatenate((angles,[angles[0]]))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, polar=True)
ax.plot(angles, stats, 'o-', linewidth=1)
ax.plot(angles, stats1, 'o-', linewidth=1)
ax.plot(angles, stats2, 'o-', linewidth=1)

ax.fill(angles, stats, alpha=0.25)
ax.fill(angles, stats1, alpha=0.25)
ax.fill(angles, stats2, alpha=0.25)

ax.legend(['2004-01-01','2008-01-01','2010-01-01'])

ax.set_thetagrids(angles*(180/np.pi), labels)
ax.grid(True)
```

```python
f2 = open('US_births_2000-2014_SSA.csv')

def genData3(f2):
    """
    Generate Data3
    
    data3 = genData3(f2)
    """
    lines2 = f2.readlines()
    data3 = defaultdict(int)
    for i in range(1,len(lines2)):
        cur = lines2[i].split(',')
        data3[int(cur[1])]+=float(cur[4])
    return data3

data3 = genData3(f2)
```

```python
f3 = open('fifa_countries_audience.csv')

def genDics(f3):
    """
    Generate hash map
    for population, tv, and gdp sharings
    
    pop,tv,gdp = genDics(f3)
    """
    lines3 = f3.readlines()
    pop,tv,gdp = defaultdict(int),defaultdict(int),defaultdict(int)
    for i in range(1,len(lines3)):
        cur = lines3[i].split(",")
        if i!=164:
            dt = [float(x) for x in cur[2:]]
            pop[cur[1]]+=dt[0]
            tv[cur[1]]+=dt[1]
            gdp[cur[1]]+=dt[2]
        else:
            dt = [float(x) for x in cur[3:]]
            pop[cur[2]]+=dt[0]
            tv[cur[2]]+=dt[1]
            gdp[cur[2]]+=dt[2]
    return pop,tv,gdp

pop,tv,gdp = genDics(f3)
```

```python
conf = list(pop)
l_pop = list(pop.values())
l_tv = list(tv.values())
l_gdp = list(gdp.values())
plt.plot(conf,l_pop)
plt.plot(conf,l_tv)
plt.plot(conf,l_gdp)
plt.legend(["Population share","Tv share","GDP share"])
plt.xlabel("Confederations")
plt.ylabel("Sharing (%)")
plt.grid()
```

```python
labels = conf
stats_pop = l_pop
angles = np.linspace(0, 2*np.pi,len(labels),endpoint=False)
stats_pop = np.concatenate((stats_pop,[stats_pop[0]]))
angles = np.concatenate((angles,[angles[0]]))

stats_tv = l_tv
stats_tv = np.concatenate((stats_tv,[stats_tv[0]]))

stats_gdp = l_gdp
stats_gdp = np.concatenate((stats_gdp,[stats_gdp[0]]))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, polar=True)
ax.plot(angles, stats_pop, 'o-', linewidth=1)
ax.fill(angles, stats_pop, alpha=0.25)
ax.set_thetagrids(angles*(180/np.pi), labels)

ax.plot(angles,stats_tv,'o-',linewidth=1)
ax.fill(angles, stats_tv, alpha=0.25)


ax.plot(angles,stats_gdp,'o-',linewidth=1)
ax.fill(angles, stats_gdp, alpha=0.25)
ax.legend(["Population share","Tv share","GDP share"])
ax.grid(True)
```

```python
def threshold(tr):
    """
    It is for threshold
    """
    img = makeNew(newImage)    
    m,n = len(img),len(img[0])
    for i in range(m):
        for j in range(n):
            if img[i][j]<tr:
                img[i][j]=0
            else:
                img[i][j]=1
    plt.imshow(img,'gray')

def makeNew(newImage):
    """
    It is for preventing origial data is changed
    """
    m,n = len(newImage),len(newImage[0])
    copy = [[0 for x in range(n)] for x in range(m)]
    for i in range(m):
        for j in range(n):
            copy[i][j] = newImage[i][j]
    return copy 
```
